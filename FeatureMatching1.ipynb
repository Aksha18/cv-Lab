```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(10,12))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses = cv2.imread('E:/Computer-Vision-with-Python/DATA/reeses_puffs.png',0)
```

```python
display(reeses)
```

```python
cereals = cv2.imread('E:/Computer-Vision-with-Python/DATA/many_cereals.jpg',0)
```

```python
display(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1=orb.detectAndCompute(reeses,None)
kp2,des2=orb.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches[2].distance
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:15],None,flags=2)
```

```python
display(reeses_matches)
```
