```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('E:/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat,cmap='gray')
```

```python
gray = np.float32(gray_flat)
```

```python
gray
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
#HARRIS CORENER DETECTION FOR REAL CHESS
real_chess = cv2.imread('E:/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
real_gray= cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(real_gray,cmap='gray')
```

```python
real_gray1=np.float32(real_gray)
```

```python
dst1 = cv2.cornerHarris(real_gray1,2,3,0,0.04)
```

```python
dst1 = cv2.dilate(dst1,None)
```

```python
real_chess[dst1>0.01*dst1.max()]=[0,255,0]
```

```python
plt.imshow(real_chess)
```
