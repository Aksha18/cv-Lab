```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse =cv2.imread('E:/Computer-Vision-with-Python/DATA/horse.jpg')
show_horse= cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)
                         
blue_bricks = cv2.imread('E:/Computer-Vision-with-Python/DATA/bricks.jpg')
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)

rainbow = cv2.imread('E:/Computer-Vision-with-Python/DATA/rainbow.jpg')
show_rainbow=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)
                         
```

```python
plt.imshow(show_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
#creating histograms
#0 for Blue Channel
hist_values =cv2.calcHist([show_bricks],channels=[0],mask = None,histSize=[256],ranges=[0,256])
```

```python
#it shows the blue contribution in the image blue bricks
plt.plot(hist_values)
```

```python
#seeing the red color in the bricks image
hist_values1= cv2.calcHist([dark_horse],channels=[0],mask= None ,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_values1)
```

```python
#SEEING THREE COLOR IN HISTOGRAM

img = blue_bricks
color = ['b','g','r']

for i ,col in enumerate(color):
     histr = cv2.calcHist([img],[i],None,[256],[0,256])
     plt.plot(histr,color=col)
     plt.xlim([0,50])
plt.title('HISTOGRAM FOR BLUE BRICKS')
    
```
