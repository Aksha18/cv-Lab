```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
real_chess=cv2.imread('E:/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess)
```

```python
flat_chess = cv2.imread('E:/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
plt.imshow(real_chess,cmap='gray')
```

```python
corners= cv2.goodFeaturesToTrack(gray_flat_chess,6,0.01,10)
```

```python
corners=np.int0(corners)
```

```python
for i in corners:
    x,y= i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners1 = cv2.goodFeaturesToTrack(gray_real_chess,100,0.01,10)
```

```python
corners1=np.int0(corners1)
```

```python
for i in corners1:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(real_chess)
```
