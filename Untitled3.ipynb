```python
def display(img,cmap='gray'):
  fig = plt.figure(figsize=(12,10))
  ax=fig.add_subplot(111)
  ax.imshow(img,cmap='gray')
```

```python
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
noida = cv2.imread('E:/Computer-Vision-with-Python/DATA/Nadia_Murad.jpg',0)
```

```python
def detect_face(img):
    face = img.copy()
    face_rect = classifier.detectMultiScale(face,scaleFactor=1.2,minNeighbors=5)
    for(x,y,w,h) in  face_rect:
        cv2.rectangle(face,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face
```

```python
result = detect_face(ruba)
```

```python
plt.imshow(result,cmap='gray')
```

```python
display(result)
```

```python



cap =cv2.VideoCapture(0)
while True:
    ret,frame = cap.read()
    
    frame= detect_face(frame)
    cv2.imshow("detected face",frame)
    
    k = cv2.waitKey(1)
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
